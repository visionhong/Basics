## DataFrame

#### 1. DataFrame생성 - dictionary 사용

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
list_dic={'state':['Ohio','Ohio','Ohio','Nevada','Nevada','Nevada'],
         'year':[2000, 2001, 2002, 2001, 2002, 2003],
         'pop':[1.5, 1.7, 1.6, 2.4, 2.9, 3.2]}

# DataFrame
'''
state,year,pop 각각의 컬럼에 해당하는 상 -> 하 방향의 한줄한줄이 시리즈라고 보면 된다.
DataFrame은 시리즈들의 결합체
시리즈는 ndarray타입으로 만들어져 있다.
'''
```

```python
dicDf = pd.DataFrame(data=list_dic)
dicDf
```

```python
print(type(dicDf['state']))
print(type(dicDf['year']))
print(type(dicDf['pop']))
print(type(dicDf.pop)) # pop이라는 method가 있어서 못씀
```

```python
data1 = {
    'name':['James','Peter','Robert','Tom'],
    'address':['NY','NY','LA','Texas'],
    'age':[33,44,55,66]
}
df1 = pd.DataFrame(data1)
df1
```
