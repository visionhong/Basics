### 1. Pivot Tables - pivot_table()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
'''
엑셀... pivot table 기능이 있다.. 이것과 동일한 기능
Pandas에서는 pivot_table() 함수를 제공
피봇함수 역할?
DataFrame의 데이타를 ReShape하는 방법중 하나.
구체적인 방법론
1.
2.
3.

'''
```

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# 기본 agg는 평균값이 나옴
pd.pivot_table(df, index=['도시'])
```

```python
df.pivot_table('인구','도시','연도') # 차례대로 값 행 열
```

```python
df.pivot_table(values='인구',columns='연도',index='도시')
```

```python
df.pivot_table(values='인구',index=['연도','지역'], aggfunc='sum')
```

```python
df.pivot_table('인구','도시','연도',margins=True,fill_value=5) # fillna
```

```python
df['인구'].mean()
# 위와 같음
```

```python
df.pivot_table(['인구'], index=['연도','도시'])
```

### 실전데이터 응용하기

```python
tipdf = pd.read_csv('../data/tips.csv')
tipdf

'''
식사대금 대비 팁의 비율이 어떤지 알수있는 컬럼 하나를 더 추가해보자
tip_pct
'''
```

```python
tipdf['tip_pct'] = round(tipdf['tip'] / tipdf['total_bill'] *100,2)
```

```python
tipdf = tipdf.dropna() # default = any
```

```python
# 데이타들에 대한 간략한 통계정보 출력
tipdf.describe()
```

```python
tipdf.info()
```

```python
tipdf.pivot_table(['tip'],index=['sex','size'])
```

```python
# 문제1. tipdf 에서 pivot 테이블을 사용해서 day라는 컬럼으로 세분화

tipdf.pivot_table(index =['day'],aggfunc='mean').round(2)
```

```python
# 문제 2. tipdf 에서 모든 데이터를 성별, 흡연여부별로 세분화... 소수점 2자리까지만 출력
tipdf.pivot_table(index=['sex','smoker'])
```

```python
'''
pivot_table() 함수 인자값 속성의 의미
index : 세분화, 그룹핑 하는 기준
columns : 컬럼지정
values - 데이타프레임을 채우고자 하는 값
'''
tipdf.pivot_table(values=['total_bill','tip'], index=['sex','day'], columns='smoker')
```

```python
tipdf.pivot_table('tip','sex','smoker')
```

```python
# 문제 3. 성별에 따른 탑의 평균비율
tipdf.pivot_table('tip_pct','sex').round(2)
```

```python
tipdf.groupby('sex')[['tip_pct']].mean().round(2)
```

```python
# 문제 4. ...groupby()로 해결
'''
각 그룹에서 가장 많은 팁과 가장 적은 팁의 차이를 디스플레이 하시오
단 사용자정의 함수를 지정... max_min_tip으로 정의해주세요
'''

def max_min_tip(x):
    a = x.min()
    b = x.max()
    return b-a

tipdf.groupby(['sex','smoker'])['tip'].agg([max_min_tip])
```

```python
tipdf.pivot_table('tip',index=['sex','smoker'], aggfunc=max_min_tip)
```
