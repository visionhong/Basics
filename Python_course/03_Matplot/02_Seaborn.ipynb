#### 2. BoxPlot

```python
# 0부터 100까지의 숫자 총 101개를 만든다.
xs = np.arange(101).astype('float64')

df = pd.DataFrame(xs,columns=['features'])
df
```

```python
# 뒤의 데이터프레임을 BoxPlot으로
sns.set_style('darkgrid')
plt.figure(figsize=(7,6)) # 그래프사이즈 지정
df.boxplot(column='features')

plt.yticks(np.arange(0,101,5))
plt.show()
```

#### 실전데이터 응용

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.figure(figsize=(4,2))
sns.boxplot(data=tips,x='total_bill')
plt.show()
```

```python
tips.sort_values(by='total_bill',ascending=False).head(10)
```

```python
# 날짜별로 음식 지출액
plt.figure(figsize=(10,20))
sns.boxplot(x='day',y='total_bill',data=tips)
plt.yticks(np.arange(0,60,1))
plt.show()
```

```python
# 요일별 흡연여부에 따른 지출비용
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips)
```

```python
### 

sns.boxplot(x='day',y='tip',hue='sex', data=tips)
```

#### 3. Implot

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# 팁을 더 많이 주는 사람은 비흡연자
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips)
```

#### 4. Heatmap

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights = flights.pivot_table(index='month',columns='year',values='passengers')
flights
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data=flights, annot=True, fmt='d') #fmt는 정수값
plt.show()
```
